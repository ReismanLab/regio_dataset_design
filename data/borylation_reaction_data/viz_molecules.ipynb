```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
```

```python
df = pd.read_csv('borylation_regio.csv')
smiles = df.Reactant_SMILES.unique()
mols = [Chem.MolFromSmiles(smi) for smi in smiles]
Draw.MolsToGridImage(mols, molsPerRow=4, subImgSize=(200, 200), maxMols=100)
```
